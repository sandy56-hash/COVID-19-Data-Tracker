```python
# %% [markdown]
# # COVID-19 Data Analysis
# ## 1 Data Loading & Exploration

# %%
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
df = pd.read_csv('country_data.csv')  # Make sure the filename matches your uploaded file

# Initial exploration
print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())

print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())

# Basic statistics
print("\nDescriptive statistics:")
display(df.describe())

# %% [markdown]
# ## 2 Data Cleaning

# %%
# Convert date-like columns if they exist (though not in current dataset)
# df['date'] = pd.to_datetime(df['date'])

# Handle missing values
df_clean = df.copy()

# Fill numeric columns with median
numeric_cols = df.select_dtypes(include=np.number).columns
df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].median())

# Verify cleaning
print("Remaining missing values:\n", df_clean.isnull().sum())

# %% [markdown]
# ## 3 Exploratory Data Analysis (EDA)
# %%
# Set visualization style 
import seaborn as sns
sns.set_theme(style="whitegrid", palette="pastel")  # Updated styling
plt.rcParams['figure.figsize'] = (12, 6)


# Top 10 countries by confirmed cases
top_countries = df_clean.sort_values('Confirmed', ascending=False).head(10)
plt.figure(figsize=(14, 6))
sns.barplot(x='Confirmed', y='Country/Region', data=top_countries, palette='viridis')
plt.title('Top 10 Countries by Confirmed Cases')
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Country')
plt.show()

# Death rate analysis
df_clean['Death_Rate'] = (df_clean['Deaths'] / df_clean['Confirmed']) * 100
high_death_rate = df_clean[df_clean['Confirmed'] > 10000].sort_values('Death_Rate', ascending=False).head(10)

plt.figure(figsize=(14, 6))
sns.barplot(x='Death_Rate', y='Country/Region', data=high_death_rate, palette='magma')
plt.title('Top 10 Countries by Death Rate (Confirmed Cases > 10,000)')
plt.xlabel('Death Rate (%)')
plt.ylabel('Country')
plt.show()

# Recovery rate analysis
df_clean['Recovery_Rate'] = (df_clean['Recovered'] / df_clean['Confirmed']) * 100
high_recovery = df_clean[df_clean['Confirmed'] > 10000].sort_values('Recovery_Rate', ascending=False).head(10)

plt.figure(figsize=(14, 6))
sns.barplot(x='Recovery_Rate', y='Country/Region', data=high_recovery, palette='plasma')
plt.title('Top 10 Countries by Recovery Rate (Confirmed Cases > 10,000)')
plt.xlabel('Recovery Rate (%)')
plt.ylabel('Country')
plt.show()

# %% [markdown]
# ## 4 Comparative Analysis by WHO Region

# %%
# Aggregate by WHO Region
region_stats = df_clean.groupby('WHO Region').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Active': 'sum'
}).sort_values('Confirmed', ascending=False)

# Calculate rates
region_stats['Death_Rate'] = (region_stats['Deaths'] / region_stats['Confirmed']) * 100
region_stats['Recovery_Rate'] = (region_stats['Recovered'] / region_stats['Confirmed']) * 100

display(region_stats)

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Confirmed Cases
sns.barplot(x='Confirmed', y=region_stats.index, data=region_stats, ax=axes[0,0], palette='viridis')
axes[0,0].set_title('Total Confirmed Cases by WHO Region')

# Deaths
sns.barplot(x='Deaths', y=region_stats.index, data=region_stats, ax=axes[0,1], palette='magma')
axes[0,1].set_title('Total Deaths by WHO Region')

# Death Rate
sns.barplot(x='Death_Rate', y=region_stats.index, data=region_stats, ax=axes[1,0], palette='plasma')
axes[1,0].set_title('Death Rate (%) by WHO Region')

# Recovery Rate
sns.barplot(x='Recovery_Rate', y=region_stats.index, data=region_stats, ax=axes[1,1], palette='cividis')
axes[1,1].set_title('Recovery Rate (%) by WHO Region')

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 5 Correlation Analysis

# %%
# Select relevant columns for correlation
corr_cols = ['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 
             'New deaths', 'New recovered', 'Deaths / 100 Cases',
             'Recovered / 100 Cases', 'Deaths / 100 Recovered',
             'Confirmed last week', '1 week change', '1 week % increase']

corr_matrix = df_clean[corr_cols].corr()

# Plot heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f',
            linewidths=0.5, linecolor='black')
plt.title('Correlation Matrix of COVID-19 Metrics')
plt.show()

# %% [markdown]
# ## 6 Outlier Detection

# %%
# Identify countries with unusual metrics
df_clean['Case_Fatality_Ratio'] = df_clean['Deaths / 100 Cases']
df_clean['Recovery_to_Fatality_Ratio'] = df_clean['Deaths / 100 Recovered']

# Countries with highest case fatality ratio
high_fatality = df_clean[df_clean['Confirmed'] > 1000].sort_values('Case_Fatality_Ratio', ascending=False).head(10)
print("Countries with highest case fatality ratio:")
display(high_fatality[['Country/Region', 'Confirmed', 'Deaths', 'Case_Fatality_Ratio']])

# Countries with unusual recovery to fatality ratio
unusual_ratio = df_clean[
    (df_clean['Confirmed'] > 1000) & 
    (df_clean['Recovered'] > 0)
].sort_values('Recovery_to_Fatality_Ratio', ascending=False).head(10)
print("\nCountries with unusual recovery to fatality ratio:")
display(unusual_ratio[['Country/Region', 'Recovered', 'Deaths', 'Recovery_to_Fatality_Ratio']])

# ## 7 Key Insights

# %%
# Calculate global totals
total_confirmed = df_clean['Confirmed'].sum()
total_deaths = df_clean['Deaths'].sum()
total_recovered = df_clean['Recovered'].sum()

print(f"""
Global COVID-19 Statistics:
- Total Confirmed Cases: {total_confirmed:,}
- Total Deaths: {total_deaths:,} ({(total_deaths/total_confirmed)*100:.2f}% of confirmed cases)
- Total Recovered: {total_recovered:,} ({(total_recovered/total_confirmed)*100:.2f}% of confirmed cases)
""")
```
